### Question 3: Which Crypto Coins are the Most Volatile?

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
#Imports necessary applications for the assignment
```

```python
crypto_data = pd.read_csv("crypto.csv")
#Imports csv file into panda library
print(crypto_data)
#Prints data frame
```

Question 1: Which of the three most popular crypto currencies is the safest to invest in?
Throughout this process I will be using Bitcoin, Ethereum and XRP to determine which would be the most stable for a new investor to put their money in. I will be determining this through volatility based on daily changes in value.

```python
crypto_data['Change in Price'] = crypto_data['Adj Close']-crypto_data['Open']
#Creates new column measuring daily change in price of crypto coins
print(crypto_data)
```

```python
pivot_table = pd.pivot_table(crypto_data, values='Change in Price', index = 'Type', aggfunc = 'describe')
#Creates a pivot table of every crypto currency in the csv's daily change in price
#Pivot table lists quartiles, counts, maximum, mean, and minimum values along with standard deviation
print(pivot_table)
```

```python
big_three = ['Bitcoin USD', 'Ethereum USD', 'XRP USD']
#isolates the top 3 most popular crypto coins for analysis
filtered_data=crypto_data[crypto_data['Type'].isin(big_three)]
"""
Creates filtered data where only the types of crypto listed in big_three list are shown
in the graph. Source: GoogleAI
"""
filtered_data.boxplot(column="Change in Price", by='Type', flierprops={'markersize':3})
#Makes Boxplot comparing the daily change in price of different crypto currencies
#Flierprops allows for changing marker size, Source: GoogleAI
plt.title('Volatility in Popular Crypto Currencies')
#Creates Title
plt.suptitle('')
#Gets rid of 'Boxplots Grouped by Type' in title. Source: StackOverflow
plt.xlabel(' ')
#Gets rid of Type label on graph
plt.ylabel('Daily Change in Coin Value (USD)')
#Adds y axis label
plt.show()
#Shows Graph
```

Analysis:
In order to determine which of the three most popular crypto coins is the safest to invest in a box plot of the daily change in price of Bitcoin, Ethereum, and XRP was made. As can be shown in the boxplots of Bitcoin, Ethereum, and XRP, by far Bitcoin has the most daily volatility, especially given the sheer amount of outliers stretching far beyond the whiskers of the plot. Since volatility measures the frequency and amount of change an investment changes, in general, Bitcoin is a risky investment compared to Ethereum and XRP. However, it should be noted that Bitcoin also has significantly greater gains in singular days with a maximum of almost $8000 gain in one day. While a coin like XRP would be advised for cautious investing, Bitcoin allows for high-risk, high-reward scenarios.

Question 4: "Does the amount of coins traded in the day affect the daily change in volume". To do so I will be using a chi-squared test with the change in price and volume columns.

```python
chi_table=pd.crosstab(crypto_data['Volume'], crypto_data['Change in Price'])
#Creates a contingency table between volume and change in price
chi2, p, dof, expected = scipy.stats.chi2_contingency(chi_table)
#Does a Chi^2 test of contingency table
print(p)
#prints p value
```

Analysis: The p-value between Volume and Change in Price is 0.0, this shows an extremely strong correlation between the volume, or the amount of coins traded in a day, and the daily change in price since the p-value is less than 0.05. This means that the chi-squared test displayed a clear trend where volume seems to directly effect the volatility of crypto coins. Thus it shows that crypto coins are influenced to a high degree by the actions of investors, hinting that the coins are not always so stable.
